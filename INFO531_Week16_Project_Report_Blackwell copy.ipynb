```python
# Predictor variables
X = df.drop("Churn", axis=1)

display(X.head())
```

```python
# Target variable
y = df["Churn"]

display(y.head())
```

- The predictor variables were assigned to a dataframe named X by removing the Churn column from the dataset. These predictor variables include customer demographic information, account details, service usage patterns, and billing information that may help predict customer churn.

### 3. Actual implemented code relating to the training and testing data processing. Provide descriptions of this work. Provide the output generated and interpret the results.

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
selected_columns = [
    "gender",
    "tenure",
    "Contract",
    "InternetService",
    "PaymentMethod",
    "MonthlyCharges",
    "TotalCharges"
]

X = df[selected_columns]
y = df["Churn"]

display(X.head())
display(y.head())
```

```python
X_encoded = pd.get_dummies(X, drop_first=True)

print("Original number of predictor columns:", X.shape[1])
print("Number of predictor columns after encoding:", X_encoded.shape[1])

display(X_encoded.head())
```

```python
from sklearn.model_selection import train_test_split
df = df.dropna()

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
print("Training Features Shape:", X_train.shape)
print("Testing Features Shape:", X_test.shape)
print("Training Target Shape:", y_train.shape)
print("Testing Target Shape:", y_test.shape)
```

- The selected predictor variables were reduced to a smaller group of columns to keep the machine learning model easier to interpret and focused on important customer demographic, billing, account, and service-related information. Categorical variables such as gender, contract type, internet service type, and payment method were converted into numerical format using one-hot encoding. This preprocessing step was necessary because machine learning models cannot process text values directly.

- After encoding, the dataset increased from 7 predictor columns to 11 predictor columns. The dataset was then split into training and testing datasets using an 80/20 split with the train_test_split() function from Scikit-learn. The output showed that the training feature dataset contained 5,634 rows and 11 columns, while the testing feature dataset contained 1,409 rows and 11 columns. The training target dataset contained 5,634 rows, and the testing target dataset contained 1,409 rows.

### 4. Provide the step-by-step code for each ML technique(s). Provide the output generated and interpret the results.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
log_model = LogisticRegression(max_iter=5000, solver="liblinear")

log_model.fit(X_train, y_train)

log_predictions = log_model.predict(X_test)

print("Logistic Regression Accuracy:", accuracy_score(y_test, log_predictions))
print(classification_report(y_test, log_predictions))
```

- The Logistic Regression model achieved an accuracy of approximately 79%. This means the model correctly predicted customer churn status for about 79% of the testing records.The model performed better at predicting customers who did not churn, shown by the higher precision, recall, and F1-score for class 0. The results for class 1, which represents customers who churned, were lower. This shows that predicting churned customers is more difficult, but the model still provides useful insight into customer churn patterns.

```python
tree_model = DecisionTreeClassifier(random_state=42)

tree_model.fit(X_train, y_train)

tree_predictions = tree_model.predict(X_test)

print("Decision Tree Accuracy:", accuracy_score(y_test, tree_predictions))
print(classification_report(y_test, tree_predictions))
```

- The Decision Tree model achieved an accuracy of approximately 73%. This means the model correctly predicted customer churn status for about 73% of the customers in the testing dataset. The results showed that the model performed better at identifying customers who stayed with the company than customers who churned. This can be seen through the higher precision, recall, and F1-score values for class 0 (non-churn customers). The lower scores for class 1 indicate that predicting churned customers is more difficult because customer behavior patterns can vary. Overall, the Decision Tree model demonstrated that customer contract type, billing information, and internet services can help identify customer churn patterns.

```python
forest_model = RandomForestClassifier(random_state=42)

forest_model.fit(X_train, y_train)

forest_predictions = forest_model.predict(X_test)

print("Random Forest Accuracy:", accuracy_score(y_test, forest_predictions))
print(classification_report(y_test, forest_predictions))
```

- The Random Forest model achieved an accuracy of approximately 75%. This means the model correctly predicted customer churn status for about 75% of the customers in the testing dataset.The results showed that the Random Forest model performed slightly better than the Decision Tree model. Similar to the other models, the Random Forest model predicted customers who stayed with the company more accurately than customers who churned. This is shown by the higher precision, recall, and F1-score values for class 0. Overall, the Random Forest model demonstrated that customer demographics, account details, internet service type, and billing information all contribute to predicting customer churn behavior.
